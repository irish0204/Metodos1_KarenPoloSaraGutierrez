```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def sucesion_fibonacci(n: int):
    numeros = np.zeros(n)
    numeros[0] = 1
    numeros[1] = 1
    for i in range(2,n):
        numeros[i] = numeros[i-1]+numeros[i-2]
    return numeros
```

```python
lista_numeros = sucesion_fibonacci(20)
fig = plt.figure(figsize=(8,3) )
ax1 = fig.add_subplot()
ax1.plot(lista_numeros)
```

```python
lista_numeros.size
```

```python
def numero_aureo(n: int):
    aureo = sucesion_fibonacci(n+1)[n]/sucesion_fibonacci(n)[n-1]
    return aureo
```

```python
aureo_calculado = numero_aureo(1000)
print(aureo_calculado)
```

```python
aureo_documentado = (1+np.sqrt(5))/2
print(aureo_documentado)
```

```python
error_porcentual = abs((aureo_calculado-aureo_documentado)/aureo_documentado)*100
print(error_porcentual,"%")
```
