```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 25
velocidad = 2
radio = 2
X = np.linspace(-4,4,n)
Y = X.copy()
```

```python
def potencial(x,y):
    phi = x*velocidad*(1-(radio**2/(x**2+y**2)))
    return phi

def derivadaX(f,x,y,h):
    der = (f(x+h,y)-f(x-h,y))/(2*h)
    return der

def derivadaY(f,x,y,h):
    der = (f(x,y+h)-f(x,y-h))/(2*h)
    return der

def getVelocities(X,Y):
    Ex = np.zeros((n,n))
    Ey = np.zeros((n,n))
    
    for i in range(0,n):
        for j in range (0,n):
            if not ((X[i]**2 + Y[j]**2) <=4):
                Ex[i,j] = derivadaX(potencial,X[i],Y[j],h=0.001)
                Ey[i,j] = -derivadaY(potencial,X[i],Y[j],h=0.001)

    return Ex,Ey
```

```python
Vx,Vy = getVelocities(X,Y)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot()
for i in range(0,X.size):
    for j in range(0,Y.size):
        ax.quiver(X[i],Y[j],Vx[i,j],Vy[i,j])
ax.set_xlabel("X")
ax.set_ylabel("Y")
```
