```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(x):
    y = np.sqrt(np.tan(x))
    
    return y
```

```python
X = np.linspace(0.1,1.1,100)
h = X[1]-X[0]
```

```python
def progDer(f,x,h):
    dx = (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
    return dx

def centralDer(f,x,h):
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d
```

```python
DXp = progDer(funcion,X,h)
DXc = centralDer(funcion,X,h)
```

```python
def derivadaTeorica(x):
    dx = (1/np.cos(x))**2/(2*np.sqrt(np.tan(x)))
    return dx
```

```python
DXt = derivadaTeorica(X)
```

```python
plt.scatter(X,DXp)
plt.scatter(X,DXc,c="red")
plt.scatter(X,DXt,c="yellow")
```

```python
#ERROR CENTRAL
plt.scatter(X,np.abs(DXc-DXt))
```

```python
#ERROR PROGRESIVA
plt.scatter(X,np.abs(DXp-DXt))
```

Ambas tienen errores nodales del orden de $10^{-4}$, sólo hay algunos puntos con errores del orden de $10^{-3}$.
