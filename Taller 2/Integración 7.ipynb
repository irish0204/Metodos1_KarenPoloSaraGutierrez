```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
n = 400
x = np.linspace(-1,1,n)
y = np.linspace(-1,1,n)
X,Y = np.meshgrid(x,y)
```

```python
vol_t = (3/4)*np.pi
vol_t
```

```python
def f(x,y):
    
    z = 1 - x**2 - y**2
    
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z) 
```

```python
f = np.vectorize(f)
```

```python
Z = f(X,Y)
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
```

```python
a = (x[0]-x[1])**2
vol = 0
for i in range(1,n):
    for j in range(1,n):
        
        p0 = (x[i],y[j])
        p1 = (x[i],y[j-1])
        p2 = (x[i-1],y[j-1])
        p3 = (x[i-1],y[j])
        
        f0 = f(p0[0],p0[1])
        f1 = f(p1[0],p1[1])
        f2 = f(p2[0],p2[1])
        f3 = f(p3[0],p3[1])        
        
        prom = (f0+f1+f2+f3)/4

        
        v = prom*a
        vol += v

vol
```
