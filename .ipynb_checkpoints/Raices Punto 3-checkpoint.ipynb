```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
X = np.linspace(-5,5,n)
```

```python
def DerivadaCentral(f,x,h=0.000001):
    der = (f(x+h)-f(x-h))/(2*h)
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def newtonRaphson(f,x1,tope=100):
    numero_veces = 0
    respuesta = 0
    while (round(f(x1),6) != 0) and (numero_veces <= tope):
        x1 = x1 - (f(x1)/DerivadaCentral(f,x1))
        numero_veces += 1
    
    if numero_veces > tope:
        respuesta = 999999
    else:
        respuesta = x1
        
    return respuesta
def funcion(x):
    f = 3*x**5 + 5*x**4 - x**3
    
    return f
```

```python
raices = []
i = 1
raiz_actual = 0
while i <= n and raiz_actual != 999999:
    x1=X[n-i]
    raiz = newtonRaphson(funcion,x1)
    
    if raiz != raiz_actual:
        raiz_actual = raiz
        raices.append(raiz)
        
    i+= 1
```

```python
raices
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot()
ax1.plot(X,funcion(X))
for r in raices:
    ax1.scatter(r,funcion(r))
```
