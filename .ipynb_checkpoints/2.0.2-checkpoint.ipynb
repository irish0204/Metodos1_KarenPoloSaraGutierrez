```python
%matplotlib notebook
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Maximos(ruta: str):
    archivo = open(ruta, "r")
    X = []
    Y = []
    
    Xmaximo = []
    Ymaximo = []

    for linea in archivo:
        valores = linea.split()
        x = float(valores[0])
        y = float(valores[1])

        X.append(x)
        Y.append(y)
    
    array_Y = np.array(Y)
    array_X = np.array(X)
    maximos = argrelextrema(array_Y, np.greater)

    for h in range(len(maximos[0])):
        i = maximos[0][h]
        Ymaximo.append(array_Y[i])
        Xmaximo.append(array_X[i])
        
    return X,Y,Xmaximo,Ymaximo
```

```python
ruta = "EstrellaEspectro.txt"
X,Y,Xmax,Ymax = Maximos(ruta)
plt.plot(X,Y)
plt.scatter(Xmax,Ymax)
plt.show()
```
