```python
from matplotlib import pyplot as plt 
import numpy as np
from mpl_toolkits.mplot3d import axes3d
```

## Punto 5

```python
N=10000
radio= 2
X = np.zeros(N)
Y = np.zeros_like(X)
Z = np.zeros_like(X)

for i in range(N):
    u=np.random.rand()
    r = radio*np.sqrt(u)
      
    theta = np.random.uniform(0,2*np.pi)
        
    h=np.sqrt(radio**2-r**2)
    z=np.random.uniform(-h,h)
        
    X[i]=r*np.cos(theta)
    Y[i] = r*np.sin(theta)
    Z[i] = z
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot( projection = '3d')

min=-2
max=2

ax.set_xlim3d(min, max)
ax.set_ylim3d(min, max)
ax.set_zlim3d(min, max)


ax.view_init(10, 60)

ax.scatter(X,Y,Z,color='b')
```

```python
def fxx(x,y,z):
    return y**2 + z**2

def fyy(x,y,z):
    return x**2 + z**2

def fzz(x,y,z):
    return x**2 + y**2
```

```python
muestraxx = np.zeros(N)
muestrayy = np.zeros(N)
muestrazz = np.zeros(N)

for i in range(N):
    muestraxx[i] = fxx(X[i],Y[i],Z[i])
    muestrayy[i] = fyy(X[i],Y[i],Z[i])
    muestrazz[i] = fzz(X[i],Y[i],Z[i])
    
Ixx = np.average(muestraxx)
Iyy = np.average(muestrayy)
Izz =np.average(muestrazz)
```
